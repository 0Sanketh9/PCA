```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale 
```

```python
wine=pd.read_csv("wine.csv")
wine
```

```python
wine['Type'].value_counts()
```

```python
# Considering only numerical data 
wine.data = wine.iloc[:,1:]
wine.data.head()
# Converting into numpy array
wine1 = wine.data.values
wine1
```

```python
wine_Normal=scale(wine1)
wine_Normal
```

```python
pca=PCA(n_components=13)
wine_pca=pca.fit_transform(wine_Normal)
wine_pca
```

```python
pca.components_ # weights
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1,color="red")
```

```python
 wine_pca[:,0:1]         #scores for PCA1 similarly can find out for others
```

```python
final_df=pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:6],columns=['PC1','PC2','PC3','PC4','PC5','PC6'])],axis=1)
final_df
```

```python
fig=plt.figure(figsize=(10,9))
sns.scatterplot(data=final_df)
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(12,8))
dendrogram=sch.dendrogram(sch.linkage(wine_Normal,'complete'))
```

```python
h_complete = AgglomerativeClustering(n_clusters=6, linkage='complete',affinity = "euclidean").fit(wine_Normal) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
wine['clust']=cluster_labels # creating a  new column and assigning it to new column 
wine
```

```python
wine.iloc[:,1:].groupby(wine.clust).mean()
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1,6):
    clf = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    clf.fit(wine_Normal)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS

plt.plot(range(1, 6), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=6, random_state=0)
y_kmeans = clf.fit_predict(wine_Normal)
```

```python
clf.inertia_
```

```python
y_kmeans
```

```python
clf.cluster_centers_
```

```python
md=pd.Series(y_kmeans)  # converting numpy array into pandas series object 
wine['clust']=md # creating a  new column and assigning it to new column 
wine
```
